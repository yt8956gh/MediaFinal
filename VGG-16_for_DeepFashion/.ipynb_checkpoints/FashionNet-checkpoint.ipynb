## Training the model for one category of features

### Preprocess data
##### 在訓練前會先預處理資料
##### 包括分別讀進訓練資料、測試資料、resize等處理

```python
# 列出要分類的類別數量(category組:50，attribute組則依attribute的個數自訂)

nb_classes = 3

# 可以定義每個class所代表的名稱，For Displaying

class_name = {
    0: 'Checked',
    1: 'Floral',
    2: 'Graphic',
    }
```

```python
# dimensions of our images

img_width, img_height = 150, 150

# train & test data path

train_data_dir = './Dataset/Train'
test_data_dir = './Dataset/Validation'

train_data = []
train_label = []
test_data = []
test_label = []
```

```python
# Simple example for loading training data. 
# loading testing data by yourself 

num_dir=0
for root, dirs, files in os.walk(train_data_dir):
    for f in files:  
        # print(os.path.join(root,f))
        img = Image.open(os.path.join(root,f))
        img = img.resize((img_width, img_height))
        img = np.array(img)
        train_data.append(img)
        train_label.append(num_dir)
    num_dir += 1
```

```python
# 將data type轉換成 numpy array

train_data = np.array(train_data)
train_label = np.array(train_label)

# 為了使用categorical_cross_entropy loss，我們需要先將label轉換成one-hot encoding

train_label = utils.to_categorical(train_label, num_classes=num_dir)

# data shape(影像數量, 長, 寬, 深度) RGB深度=3、灰階=1

print('training data shape : ', train_data.shape)
print('training label shape :', train_label.shape)
```

```python
# 使用keras VGG-16架構，並自行接上Fully-connect layer (Dense) 來進行分類
# 最後一層Fully-connect layer參數量則為分類數量

vgg16 = VGG16(include_top=False, weights='imagenet')
x = vgg16.output
x = GlobalAveragePooling2D()(x)
x = Dense(1000, activation='relu')(x)
predict = Dense(3, activation='softmax')(x)

#Define input and output of the model

model = Model(inputs=vgg16.input, outputs=predict)
```

```python
# Using Adam optimizer with lower learning rate
# 針對不同問題可以使用不同的Loss function訓練

adam=optimizers.Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=1e-08, decay=0.0)
# sgd = optimizers.SGD(lr=0.1, decay=1e-6, momentum=0.9, nesterov=True)
model.compile(loss = 'categorical_crossentropy',
              optimizer = adam,
              metrics=['accuracy'])
```

```python
# Training the model for 5 epochs

model.fit(
          x=train_data ,
          y=train_label,
          batch_size = 32,
          nb_epoch = 5,
          validation_split = 0.1,
          verbose = 1
)
```

```python
# 儲存訓練完成後的model權重，即完成model訓練

model.save_weights('./models/vgg_weights_pattern.h5')
```

```python
# testing model 需準備好test_data及test_label

loss, accuracy = model.evaluate(x=test_data, y=test_label, batch_size=32, verbose=1)

print("Validation: accuracy = %f  ;  loss = %f" % (accuracy, loss))
```

### Similarity Function

```python
# 利用測試資料的feature與我們收集的專家穿搭資料的feature
# 求出和測試資料最相似的的專家穿搭照

def similarity(feature_data,inp_feature_data):
    num_samp=inp_feature_data.size
    sim_score={}
    for i in range(len(feature_data)):
        score=0
        score_m= inp_feature_data - feature_data[i]
        score = num_samp-np.count_nonzero(score_m)
        sim_score[i]=score
    return sim_score
```

```python
# 展示相似度較高的前幾名

similarities=similarity(feature_data,inp_feature_data)
sorted_similarities = sorted(similarities.items(), key=operator.itemgetter(1),reverse=True)
#print (sorted_similarities)
num_reco=30
num_data=feature_data.size
for i in range(num_reco):
    ind = sorted_similarities[i][0]
    print("Score : ", sorted_similarities[i][1])
    show_sample(data_images[ind])
```
